## うつや不安についての探索的データ分析
データは、うつ病不安ストレス スケール (DASS) のオンライン バージョンで収集されたものです。データについては、http://www2.psy.unsw.edu.au/dass/ を参照してください。  
今回扱うデータについては、同一ディレクトリのDATA配下に格納しています。  
  
この調査は誰でも参加できるものみたいです。個人情報について付加されないアンケートで、回答者のうち、本データを調査用に利用して良いものかどうか、正解な回答をしているかどうかについて同意したものをオープンデータにしているようです。  
  
これらデータは 2017年から 2019年に収集されたとのこと。  


アンケート調査の項目には、以下42項目が含まれています。
  

各項目は、回答者ごとにランダムな順序で一度に1つずつ表示され、過去1週間にどのくらいの頻度でそうであっ
たかを、 4つの評価スケールのどれに当てはまるかを回答させるものになります。

```
1 = まったく当てはまらなかった 
2 = 少し当てはまる、または時々当てはまる 
3 = ある程度当てはまる 
4 = 当てはまる
```

これら回答については、変数A (Q1Aなど) に格納されます。  
また、その質問に回答するのにかかった時間 (ミリ秒単位) (E) と調査におけるその質問の位置 (I) も記録されました。  



上記アンケート項目以外に、補助的な属性情報を取得しているようです。
  
TIPIなる指標が10個あるみたいです。  
The Ten Item Personality Inventory was administered (see Gosling, S. D., Rentfrow, P. J., & Swann, W. B., Jr. (2003). A Very Brief Measure of the Big Five Personality Domains. Journal of Research in Personality, 37, 504-528.):

[TEN ITEM PERSONALITY MEASURE (TIPI)](https://gosling.psy.utexas.edu/scales-weve-developed/ten-item-personality-measure-tipi/)

```
TIPI1 外向的、熱狂的
TIPI2 クリティカル、口論好き
TIPI3 頼りになる、自制心がある
TIPI4 不安で、動揺しやすい
TIPI5 新しい経験に対してオープンで、複雑
TIPI6 行動に規則性があり、静か
TIPI7 共感的、温かい
TIPI8 まとまりがなく、不注意
TIPI9 落ち着いていて、情緒的に安定
TIPI10 型にはまった、非創造的
```

TIPI 項目はそれぞれについて、「私は自分自身を次のように考えています:」と自己評価するものみたいです。
```
1 = 強くそう思わない
2 = どちらかといえばそう思わない
3 = 少しそう思わない
4 = どちらともいえない
5 = 少しそう思う
6 = どちらかといえばそう思う
7 = 強くそう思う
```

````python
def correlation_ratio(categories, measurements):
  fcat, _ = pd.factorize(categories)
  cat_num = np.max(fcat)+1
  y_avg_array = np.zeros(cat_num)
  n_array = np.zeros(cat_num)
  for i in range(0,cat_num):
      cat_measures = measurements[np.argwhere(fcat == i).flatten()]
      n_array[i] = len(cat_measures)
      y_avg_array[i] = np.average(cat_measures)
  y_total_avg = np.sum(np.multiply(y_avg_array,n_array))/np.sum(n_array)
  numerator = np.sum(np.multiply(n_array,np.power(np.subtract(y_avg_array,y_total_avg),2)))
  denominator = np.sum(np.power(np.subtract(measurements,y_total_avg),2))
  if numerator == 0:
      eta = 0.0
  else:
      eta = np.sqrt(numerator/denominator)
  return eta
````
