### 『違い』って、何？

今回のように年齢や性別、婚姻歴などの属性データが与えられているアンケートでは、
属性値ごとに『違い』が現れる質問項目について興味があります。
例えば男性と女性で結果の差が大きいような質問項目は、
男女の『違い』を理解するために有効な質問項目といえます。

* 層別対象が質的変数(カテゴリー変数)を持つ場合、質的変数の値によって分割したデータの中に、
統計学的に有意な差(有意差)があるかどうかを分析します。これは10段階尺度など量的変数とみなせる尺度の場合には
通常t検定と呼ばれる検定手法を用います。
* 『違い』についてシンプルに差分を取るのではなく、なぜ有意差を検出するかというと、
サンプルサイズの影響を考慮したいからです。例えば男性の平均値が8点で女性の平均値が6点の質問項目
だったとしても、男性の数が1人、女性の数が3人のような少数データであれば、この差は意味がある差とは
言えません。シンプルに差分を取れば２点の差がありますが、統計的には有意な差はない、と言えます
* また有意差が検出された場合にも差の絶対値が小さい場合には、統計学的に意味があったとしても「無視しても構わない」ものとなります。※実際には無視できないです(近似するデータ分布として把握しておく、理解しておく必要はあります)。では、この統計学的に意味があるけど、差があるかどうかをどのように評価すれば良いかというと、ここに**効果量**という概念があります。分析方法によって効果量の定義は異なります。

![デモ2](https://drive.google.com/uc?id=1Ve8stT5VjRwl1H2mvWfO_B2kIXPQHTLY)
* 今回はt検定をベースに分析してきますので効果量には、0.2以上のもの注目すると良いです。

|  分析名  |  効果量の指標 | 効果量(大) |  効果量(中) | 効果量(小) |
| ---- | ---- | ---- | ---- | ---- |
|  相関分析  |  d  |  0.5  |  0.3  |  0.1  |
|  重回帰分析  |  R^2  |  0.26  |  0.13  |  0.02  |
|  t検定  |  d  |  0.8  |  0.5  |  0.2  |
|  一次元分散分析  |  η^2  |  0.14  |  0.06  |  0.01  |
|  多重比較  |  r  |  0.5  |  0.3  |  0.1  |
|  カイ二乗検定  |  φ, V  |  0.5  |  0.3  |  0.1  |
| ノンパラメトリック検定 |  r  |  0.5  |  0.3  |  0.1  |

* 結論としては質的データによる層別では「統計学的に有意な差」がありかつ「効果量」がガイドの値より大きいものを検出するというアプローチによって、結果をスクリーニング(分類、ふるいわけ)します。



【補足】 [統計学的に有意」とは何を意味しているのですか？](https://www.yodosha.co.jp/jikkenigaku/statistics/q1.html)  
【補足】 [NRLナレッジサイト p値について](https://www.nri.com/jp/knowledge/glossary/lst/alphabet/p_value)

````python
# 効果量(cohen d：cohen_d(x,y))
from numpy import std, mean, sqrt

#correct if the population S.D. is expected to be equal for the two groups.
def cohen_d(x,y):
  nx = len(x)
  ny = len(y)
  dof = nx + ny - 2
  return abs(mean(x) - mean(y)) / sqrt(((nx-1)*std(x, ddof=1) ** 2 + (ny-1)*std(y, ddof=1) ** 2) / dof)
````

t検定の結果についてはp値 < 0.001に対して＊＊＊,  
0.001<= p値 <0.01に対して＊＊,  
0.01<= p値 <0.05に対して＊,   
0.05<= p値 <0.1に対して † を  
p値が0.1以上の場合には非有意(ns：non significant)と評価  

効果量については0.2より大きい場合に'Small',  
0.5より大きい場合に'Medium',  
0.8より大きい場合に'Large'を出力します。  
0.2以下の場合には'Very Small'として出力します。

````python
def creat_stat_report(xa,xb):
  report_list = []
  t, p = stats.ttest_ind(xa, xb, equal_var=False)
  MU = xa.mean()-xb.mean()
  Cd = cohen_d(xa,xb)
  report_list.append(round(xa.mean(),3))
  report_list.append(round(xb.mean(),3))
  report_list.append(round(MU,3))
  report_list.append(round(p,3))
  str1 = ''
  str2 = ''

  # pval evalustion：p値の評価
  if p < 0.001:
      str1 = '***'
  elif p < 0.01:
      str1 = '**'
  elif p < 0.05:
      str1 = '*'
  elif p < 0.1:
      str1 = '†'

  # cohen d evalustion：効果量の評価
  if Cd > 0.8:
      str2 = 'Large'
  elif Cd > 0.5:
      str2 = 'Medium'
  elif Cd > 0.2:
      str2 = 'Small'
  else:
      str2 = 'Very Small'

  report_list.append(str1)
  report_list.append(round(Cd,3))
  report_list.append(str2)
  
  return(report_list)
````

````python
# レポートの形式を定義します
pre_list = []
for j in range(0,u):
    pre_list.append(j)
df_report = pd.DataFrame(columns=['ave','ave_other','ave_diff','pval','s/ns','d','d_eval'],index = pre_list)

# 形を確認しておきましょう
df_report
````

````python
for j in range(0,u):
  Col1 = data_dass_main['gender'].name
  Col2 = data_dass_main['Q1A'].name 
  xa = data_dass_main[data_dass_main[Col1]==j][Col2]
  xb = data_dass_main[data_dass_main[Col1]!=j][Col2]
  df_report.iloc[j,:] = creat_stat_report(xa,xb)
df_report
````

````python
data_dass_main_example1 = data_dass_main[['Q5A','Q41A','gender', 'age']]
````

````python
# カテゴリー変数(質的変数)の、列の位置を取得しておく
category_columns_position = [ data_dass_main_example1.columns.get_loc(col) for col in list(data_dass_main_example1.select_dtypes('object').columns) ]
category_columns_position
````

KPrototypeライブラリを用いてクラスモデルをオブジェクトとして呼び出し、実行します。  
前回Kmeanを利用したときはクラスター数をランダムに決めてましたが、今回はエルボー法を用いてクラスター数を決めることにします。  
クラスタリングの当てはまりの良さを表す指標として、KPrototypeでは、属性`.cost_`を使います。  
ちなみにscikit learnのKMeansの場合は、属性`.inertia_`です

````python
import tqdm
````

````python
list_n_cluster = []
list_cost = []
for _n in tqdm.tqdm(range(1,30,1)):
  try:
    kprototype = KPrototypes(n_jobs = -1, n_clusters = _n, init = 'Huang', random_state = 0)
    kprototype.fit_predict(data_dass_main_example1, categorical = category_columns_position)
    list_n_cluster.append(_n)
    list_cost.append(kprototype.cost_)
  except:
      break

plt.plot(list_n_cluster, list_cost)
plt.xlabel('n_cluster')
plt.ylabel('cost')
````
