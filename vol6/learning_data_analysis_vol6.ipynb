アンケートの質問項目を、回答結果の類似度をもとにグルーピングするためのテクニックとして、因子分析を学ぶ回となります。アンケート分析という意味では、ポピュラーな基本的なテクニックになります。対象とするデータは、第4回から扱っているDASSのデータになります。39775人×172項目のデータですが、質問項目そのものは４２項目になります。
因子分析を使うと、類似性の強い質問項目をまとめてグルーピングすることができます。  
また、グループに共通する要因を考えることで、データを生成している潜在因子を推定することができます。  
これによってアンケートデータの理解をスムーズに進めることができるようになります。  
例えば42項目の質問を、傾向が似た質問項目をまとめることによって、例えば4つのグループにまとめることができる場合、単純に捉えると42項目から４つのグループを理解すればよいので、データ理解のための作業負担が軽くなります。

因子分析のためのライブラリは、`factor_analyzer`を使います。  
[FactorAnalyzerの公式ドキュメント](https://factor-analyzer.readthedocs.io/en/latest/factor_analyzer.html)  

`"因子分析に、最小残差法（ミンレス法）または最尤法を使い、オプションでPromax（プロマックス法）、Varimax(バリマックス回転)が使えるクラスです。"`

**因子数の目安を出力する**  
まずは因子数の目安となる値を求めるため、データの[固有値・固有ベクトルを出力](https://www.geisya.or.jp/~mwm48961/linear_algebra/eigenvalue2.htm)します。  
固有ベクトル・固有値は、統計学においては例えば、主成分分析という形で利用されています。  

主成分分析については本データ分析入門でも使いましたが、例えば
「変数が3つ以上ある高次元のデータに対して、より低い次元でデータのばらつきを説明する」手法になります。
一般的に、ヒトは平面で描画可能な2次元データを解釈するのは得意ですが、4次元以上のデータを図示・解釈するのは得意ではありません。  
この際に次元を落として、2次元データとして解釈可能な形にするのが主成分分析で、その際の演算に固有ベクトル・固有値を用いています。  
  

今回は、値が１以上の固有値の数を数えます。  
現段階ではここで求めた固有値という複数の値（配列）のうち、  
今回は値が１以上の固有値の個数が因子数の目安となる、程度の理解にとどめておいてください。



````python
factoranalyser_object = FactorAnalyzer()
factoranalyser_object.fit(data_dass_main_qa)

# 固有値(ev)・固有ベクトル(v)をチェック
eval, values = factoranalyser_object.get_eigenvalues()
````

````python
# 因子数の目安を出力
print("因子数の目安：",len(eval[eval > 1]))
````

上記スクリプト実行結果から今回の場合、値が１以上の固有値の数が４個であることが確認できます。  
このことは、因子数の目安として４つ程度を考えると解釈しやすいということを表しています。  
ただし、この値はあくまでも目安です。  
  
先行する理論や経験から考えられる妥当な因子数があれば、
そちらを採用すべきで因子２個や因子３個の場合の方が結果が解釈し易ければ、解釈しやすい結果を採用して
問題ありません。
  
因子数に正解はありません。一方、因子分析の目的はアンケート結果をより深く理解して、洞察（インサイト）を得ることなので、目的に最も合致した結果を使用すべきでしょう。

固有値の値を高い順にプロットした図を[スクリープロット（scree plot）](https://bellcurve.jp/statistics/glossary/5503.html#:~:text=%E3%81%82%E3%82%8B%E6%88%90%E5%88%86%E3%82%84%E5%9B%A0%E5%AD%90%E3%81%AB,%E6%B1%BA%E5%AE%9A%E3%81%99%E3%82%8B%E6%96%B9%E6%B3%95%E3%81%8C%E3%81%82%E3%82%8B%E3%80%82)と呼びます。  
スクリープロットからも値が１以上の固有値の数を確認することができます。  
またスクリープロットの「カーブ」の変化から因子数を定める方法もあるため、  
因子数を定めるにあたっては、スクリープロットを用います。
(今回は、「固有値が１以上」という基準のみで、進めていきます）

KPrototypeライブラリを用いてクラスモデルをオブジェクトとして呼び出し、実行します。  
前回Kmeanを利用したときはクラスター数をランダムに決めてましたが、今回はエルボー法を用いてクラスター数を決めることにします。  

**エルボー法**

```
クラスタごとのSSE値をプロットした図（それだけ）
SSE値が"ヒジ"のようにガクンと曲がった点（SSE値の低下がサチる場所）が最適なクラスター数とみなす
```

クラスタリングの当てはまりの良さを表す指標として、KPrototypeでは、属性`.cost_`を使います。  
ちなみにscikit learnのKMeansの場合は、属性`.inertia_`です

````python
import tqdm
````

````python
list_n_cluster = []
list_cost = []
for _n in tqdm.tqdm(range(1,30,1)):
  try:
    kprototype = KPrototypes(n_jobs = -1, n_clusters = _n, init = 'Huang', random_state = 0)
    kprototype.fit_predict(data_dass_main_example1, categorical = category_columns_position)
    list_n_cluster.append(_n)
    list_cost.append(kprototype.cost_)
  except:
      break

plt.plot(list_n_cluster, list_cost)
plt.xlabel('n_cluster')
plt.ylabel('cost')
````
