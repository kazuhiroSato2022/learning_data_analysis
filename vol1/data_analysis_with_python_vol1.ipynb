### 統計量の可視化について

**全部重要な知識ですが、正直ボリュームがすごい。。。勉強会のときには事細かく説明はしません。。。**

統計分析は、データセット内の変数が相互にどのように関係し、  
それらの関係が他の変数にどのように依存するかを理解するプロセスです。  
データが適切に視覚化されると、人間の視覚システムは関係を示す傾向とパターンを見ることができるため、  
視覚化はこのプロセスのコアコンポーネントになります。
  
このチュートリアルでは、3つのSeaborn関数について説明します。  
最も使用するのは```relplot()```です。これは、散布図と線図の2つの一般的なアプローチを使用して統計的関係を視覚化するための関数です。  
```relplot()```を、次の2つの関数を使って、1つのグラフとして組み合わせます。  

```scatterplot()```
```lineplot()```
  
のちに説明するように、これらの関数は単純で理解しやすいデータ表現を使用しているにもかかわらず、  
複雑なデータセット構造を表すことができるため、非常に効果的です。  
それらは、色相、サイズ、およびスタイルのセマンティクスを使用して最大3つの追加の変数を  
マッピングすることによって、拡張性の高い2次元グラフィックスをプロットできます。

#### 集約と不確実性の表現
より複雑なデータセットでは、x変数の同じ値に対して複数の測定値があります。  
seabornのデフォルトの動作はx、平均と平均の周りの95％信頼区間をプロットすることにより、各値で複数の測定値を集計できます。

````python
fmri = sns.load_dataset("fmri")
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri);
````

````python
type(fmri)
````

信頼区間は、ブートストラップを使用して計算されます。  
これは、大規模なデータセットでは時間がかかる可能性があります。したがって、それらを無効にできます。

````python
sns.relplot(x="timepoint", y="signal", ci=None, kind="line", data=fmri);
````

````python
sns.relplot(x="timepoint", y="signal", kind="line", ci="sd", data=fmri);
````

集計を完全にオフにするには、estimatorパラメーターをNoneに設定します。  
データが各ポイントで複数の観測値を持っている場合、誤った表現が生じる可能性があります。

````python
sns.relplot(x="timepoint", y="signal", estimator=None, kind="line", data=fmri);
````

## データセットの分布を視覚化する

データのセットを処理するとき、たいていの場合、最初に行うことは、変数がどのように分布しているかを理解することです。  
一変量分布と二変量分布を調べるためのseabornのツールのいくつかを、簡単に紹介します。

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
````

````python
sns.set(color_codes=True)
````

### 一変量分布のプロット

seabornの1変量分布を簡単に確認する最も便利な方法はdistplot()関数です。  
デフォルトでは、これはヒストグラムを描画し、カーネル密度推定（KDE）を返します。  
  
※カーネル密度推定  
ある母集団の標本のデータが与えられたとき、カーネル密度推定を使えばその母集団のデータを外挿できること

100個のランダムな値を生成したときに、その分布を、母集団を推定しつつ大まかな分布を曲線で表現してくれます。

````python
x = np.random.normal(size=100)
sns.distplot(x);
````

### ヒストグラム

ヒストグラムはよく知られているものであり、hist関数は既にmatplotlibに存在します。  
ヒストグラムは、データの範囲に沿ってビンを形成し、バーを描画して各ビンに含まれる観測値の数を示すことにより、データの分布を表します。

これを説明するために、密度曲線を削除してラグプロットを追加します。ラグプロットは、各観測で小さな垂直な棒を描画します。  
rugplot()関数を使用してラグプロット自体を作成できますが、distplot()次の場所でも使用できます。

````python
sns.distplot(x, kde=False, rug=True);
````

### カーネル密度推定

カーネル密度の推定はあまり馴染みがないかもしれませんが、分布の形状をプロットするための有用なツールになる場合があります。  
ヒストグラムと同様に、KDEプロットは、一方の軸の観測密度を他方の軸に沿った高さでエンコードします。

````python
sns.distplot(x, hist=False, rug=True);
````

KDEの描画は、ヒストグラムの描画よりも計算的に複雑です。  
何が起こるかというと、各観測値はまず、その値を中心とするガウス曲線に置き換えられます。

````python
x = np.random.normal(0, 1, size=30)
bandwidth = 1.06 * x.std() * x.size ** (-1 / 5.)
support = np.linspace(-4, 4, 200)

kernels = []
for x_i in x:

    kernel = stats.norm(x_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color="r")

sns.rugplot(x, color=".2", linewidth=3);
````

次に、これらの曲線を合計して、サポートグリッドの各ポイントでの密度の値を計算します。  
結果の曲線は、閉曲線のその面積が1になるように正規化されます。

````python
from scipy.integrate import trapz
density = np.sum(kernels, axis=0)
density /= trapz(density, support)
plt.plot(support, density);
````

このkdeplot()関数をseabornで使用すると、同じ曲線になることがわかります。  

````python
sns.kdeplot(x, shade=True);
````

### 二変量分布のプロット

2つの変数の2変量分布を視覚化することも役立ちます。  
Seabornでこれを行う最も簡単な方法は、jointplot()関数を使用することです。  
この関数は、2つの変数間の二変量（または結合）関係と、それぞれの単変量（または限界）分布の両方を別々の軸に示すマルチパネル図を作成します。

````python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])
````

#### 散布図

````python
sns.jointplot(x="x", y="y", data=df);
````

#### Hexbinプロット
ヒストグラムの2変量アナログは、六角形のビンに含まれる観測値のカウントを示すため、「hexbin」プロットとして知られています。  
このプロットは、比較的大きなデータセットで最もよく機能します。  

````python
x, y = np.random.multivariate_normal(mean, cov, 1000).T
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="hex", color="k");
````

上記のカーネル密度推定手順を使用して、2変量分布を視覚化することもできます。

````python
sns.jointplot(x="x", y="y", data=df, kind="kde");
````

2変量密度をより抽象的に表現したい場合は、単純に等高線レベルの数を増やすことができます。

````python
f, ax = plt.subplots(figsize=(6, 6))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(df.x, df.y, cmap=cmap, n_levels=60, shade=True);
````

jointplot()関数が使用する、JointGrid数字を管理すること。これがポイントになります。  
  
また柔軟性を高めるために、JointGrid直接使用して図を描きたい場合があります。  
プロット後jointplot()にJointGridオブジェクトを返します。  
これを使用して、レイヤーを追加したり、ビジュアライゼーションの他の側面を微調整したりできます。

````python
g = sns.jointplot(x="x", y="y", data=df, kind="kde", color="m")
g.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$X$", "$Y$");
````

### 線形回帰モデルを描画する関数

主要な関数が、regplot()およびlmplot()があります。  
これらの機能としては、回帰によって決定される線形関係を視覚化するために使用されます。  

````python
sns.regplot(x="total_bill", y="tip", data=tips);
````

````python
sns.lmplot(x="total_bill", y="tip", data=tips);
````

Figureの形状が異なることを除いて、結果のプロットは同じであることに注意してください。  
  
現段階として、知っておくべきもう1つの主な違いregplot()は、  
xとy変数を、単純なnumpy配列、pandas Seriesオブジェクト、またはにDataFrame渡されたpandas オブジェクト内の変数への参照として、  
さまざまな形式で受け入れることdataです。  
  
対照的に、lmplot()は、xおよびy変数は文字列として指定する必要があります。  

変数の1つが離散値を取る場合、線形回帰を近似することは可能ですが、  
この種類のデータセットによって生成される単純な散布図は、多くの場合最適ではありません。

````python
sns.lmplot(x="size", y="tip", data=tips);
````

1つの解決策として、
ランダムなノイズ（「jitter」）を離散値に追加して、それらの値の分布をより明確にすることです。  
  
jitterは散布図データにのみ適用され、回帰直線の適合自体には影響しないことに注意してください。

````python
sns.lmplot(x="size", y="tip", data=tips, x_jitter=.05);
````

2番目の解決策は、信頼区間と共に中心傾向の推定値をプロットするために、各離散ビンの観測を折りたたむことです。

````python
sns.lmplot(x="size", y="tip", data=tips, x_estimator=np.mean);
````

### さまざまな種類のモデルのフィッティング

````python
anscombe = sns.load_dataset("anscombe")
````

````python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'I'"),
           ci=None, scatter_kws={"s": 80});
````

````python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           ci=None, scatter_kws={"s": 80});
````

高次の関係のこれらの種類の存在下では、  
lmplot()およびregplot()データセット内の非線形トレンドの簡単な種類を探索するために、多項式回帰モデルを当てはめることができます。

````python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           order=2, ci=None, scatter_kws={"s": 80});
````

一方で、何らかの理由で逸脱する「異常値」の観察によって、別の問題が発生します。

````python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"),
           ci=None, scatter_kws={"s": 80});
````

外れ値が存在する場合、異なる損失関数を使用して比較的大きな残差をダウンウェイトするロバスト回帰を当てはめると便利です。  
  
※ロバスト回帰  →外れ値が存在する場合に、最小二乗法やその他手法以外で影響を軽減するもので、  
各データ点に重みを割り当てることによって機能します。  
重み付けは、"反復重み付け最小二乗法" と呼ばれる過程を使用して自動的かつ反復的に行われます  
  
下記、モジュールエラー出る場合は、pipかcondaでインストールしてください
>ModuleNotFoundError: No module named 'statsmodels'  
```pip install statsmodels```

````python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"),
           robust=True, ci=None, scatter_kws={"s": 80});
````

バイナリデータ(0, 1)にも、単純な線形回帰が適用できます(笑)

````python
tips["big_tip"] = (tips.tip / tips.total_bill) > .15
sns.lmplot(x="total_bill", y="big_tip", data=tips,
           y_jitter=.03);
````

ロジスティック回帰(線形回帰モデルの一つ)を当てはめると

````python
sns.lmplot(x="total_bill", y="big_tip", data=tips,
           logistic=True, y_jitter=.03);
````

ロジスティック回帰推定は、単純回帰よりもかなり計算量が多い（これはロバスト回帰にも当てはまります）ことに注意してください。  
回帰直線の周囲の信頼区間はブートストラップ法を使用して計算されるため、これをオフにすると高速になる場合があります。  
  
※ブートストラップ法 →母集団の推定量（分散など）の性質を、近似分布にしたがって標本化したときの性質を計算することで推定する手法
